d) generacja danych o zadanym rozkładzie: ile punktów, zakres, wymiarowość, rozkład, liczba klas,
wizualizacja

```python
mean = 0
std_dev = 1
# ile punktów
size1 = 10000
# rozkład
normal_distribution1D = np.random.normal(loc=mean, scale=std_dev, size=size1)
# liczba klas
bins = 50
# wizualizacja
plt.hist(normal_distribution1D, bins=bins)
plt.show()
```

```python
# wymiarowość
size2 = 2
normal_distribution2D = np.random.normal(loc=mean, scale=std_dev, size=(size1, size2))
plt.hist2d([x[0] for x in normal_distribution2D], [x[1] for x in normal_distribution2D], bins=bins)
plt.show()
```

```python
# zakres
lower, upper = 3.5, 6
# rozkład normalny, ograniczony od 3.5 do 6
X = stats.truncnorm(
    (lower - mean) / std_dev, (upper - mean) / std_dev, loc=mean, scale=std_dev)

plt.hist(X.rvs(10000))
plt.show()
```

```python
mean, std_dev = 5, 0.7
X = stats.truncnorm(
    (lower - mean) / std_dev, (upper - mean) / std_dev, loc=mean, scale=std_dev)

plt.hist(X.rvs(10000))
plt.show()
```

```python
# Expected number of events occurring in a fixed-time interval
lam = 5
poisson_distribution = np.random.poisson(lam=lam, size=size1)
plt.hist(poisson_distribution, bins=bins)
plt.show()
```

e) zapis podzakresu danych do pliku

```python
breast_cancer_features_ds = pd.DataFrame(breast_cancer_ds.data, columns=breast_cancer_ds.feature_names)
breast_cancer_features_ds.head()
```

```python
breast_cancer_features_subset_ds = breast_cancer_features_ds[breast_cancer_features_ds['mean texture'] > 20].copy()
starting_row, ending_row = 50, 150
columns = ['mean texture', 'worst texture']
breast_cancer_features_subset_ds = breast_cancer_features_subset_ds.loc[starting_row:ending_row, columns]
breast_cancer_features_subset_ds.to_csv('breast_cancer_features_subset.csv', index=False)
```
