```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets as ds
import pandas as pd
```

a) rozk≈Çady zmiennych: histogram, diagram rozrzutu

- histogram

```python
size = 10000
normal_distribution1D = np.random.normal(size=size)
bins = 50
plt.hist(normal_distribution1D, bins=bins)
plt.show()
```

- macierz rozrzutu

```python
size = 1000
normal_distribution2D = np.random.normal(0, 0.1, size=(size, size))
x = [pair[0] for pair in normal_distribution2D]
y = [pair[1] for pair in normal_distribution2D]
plt.scatter(x, y)
plt.show()
```

b) macierz kowariancji

```python
diabetes_ds =ds.load_diabetes()
diabetes_features_df = pd.DataFrame(diabetes_ds.data, columns=diabetes_ds.feature_names)
diabetes_target_df = pd.DataFrame(diabetes_ds.target, columns=['target'])
```

```python
diabetes_features_df.columns
```

```python
labels = ['age', 'bmi', 'target']
data = np.array([diabetes_features_df[labels[0]], diabetes_features_df[labels[1]], diabetes_target_df[labels[2]]])
cov = np.cov(data)
cov
```

```python
sns.heatmap(cov, annot=True, xticklabels=labels, yticklabels=labels, fmt='g')
plt.show()
```
