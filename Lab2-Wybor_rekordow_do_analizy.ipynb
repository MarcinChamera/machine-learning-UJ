```python
import pandas as pd
import numpy as np
import sklearn.datasets as ds
```

```python
breast_cancer_ds = ds.load_breast_cancer()
features_df = pd.DataFrame(breast_cancer_ds.data, columns=breast_cancer_ds.feature_names)
targets_df = pd.DataFrame(breast_cancer_ds.target, columns=['target'])
df = features_df.join(targets_df)
df
```

a) w zakresie od-do

```python
start_record = 20
end_record = 100
df.loc[start_record:end_record, ]
```

b) losowo

```python
number_of_records = 100
df.sample(n=number_of_records)
```

```python
fraction = 0.2
df.sample(frac=fraction)
```

c) losowo z uwzględnieniem klas

```python
df['target'].unique()
```

```python
first_class_df = df[df['target'] == 0]
second_class_df = df[df['target'] == 1]
print('class 0:', len(first_class_df), '\nclass 1:', len(second_class_df),
      '\nratio class 0 / class 1:', len(first_class_df)/len(second_class_df))
```

```python
first_class_sample_df = first_class_df.sample(frac=0.2)
second_class_sample_df = second_class_df.sample(frac=0.2)
print('class 0:', len(first_class_sample_df), '\nclass 1:', len(second_class_sample_df),
      '\nratio class 0 / class 1:', len(first_class_sample_df)/len(second_class_sample_df))
first_class_sample_df.append(second_class_sample_df) 
```
