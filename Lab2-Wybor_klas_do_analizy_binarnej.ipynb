```python
import pandas as pd
import sklearn.datasets as ds
```

```python
iris_ds = ds.load_iris()
features_df = pd.DataFrame(iris_ds.data, columns=iris_ds.feature_names)
targets_df = pd.DataFrame(iris_ds.target, columns=['target'])
df = features_df.join(targets_df)
print(iris_ds.target_names)
df.head()
```

```python
# zwraca rekordy w losowej kolejności
df = df.sample(frac=1).reset_index(drop=True)
```

```python
df['target'].unique()
```

a) dwie klasy z wielu

```python
first_class = 0
second_class = 1
two_classes_df = df[(df['target'] == first_class) | (df['target'] == second_class)].reset_index(drop=True)
print(two_classes_df['target'].unique())
two_classes_df.head(10)
```

b) jedna klasa przeciw pozostałym

dla wybranego zestawu danych (iris) potrzebne będą 3 klasyfikatory binarne (liczba klasyfikatorów binarnych równa się liczbie występujących klas)

dlatego też, należy utworzyć 3 podzbiory, gdzie w każdym będzie rozważana przynależność tylko do jednej, konkretnej klasy

```python
setosa_df = df.rename(columns={'target' : 'setosa'})
setosa_df['setosa'] = setosa_df['setosa'].apply(lambda x: 1 if x == 0 else -1)

versicolor_df = df.rename(columns={'target' : 'versicolor'})
versicolor_df['versicolor'] = versicolor_df['versicolor'].apply(lambda x: 1 if x == 1 else -1)

virginica_df = df.rename(columns={'target' : 'virginica'})
virginica_df['virginica'] = virginica_df['virginica'].apply(lambda x: 1 if x == 2 else -1)

setosa_df.head(10)
```

```python
virginica_df.head(10)
```

c) każde dwie klasy w parach

dla wybranego zestawu danych (iris) potrzebnych będzie N * (N-1)/2 klasyfikatorów binarnych (czyli dla N=3 potrzeba 3*(3-1)/2=3)

dlatego też, należy utworzyć 3 podzbiory, gdzie w każdym będzie rozważana przynależność tylko do jednej, konkretnej klasy

```python
setosa = 0
versicolor = 1
virginica = 2
setosa_versicolor_df = df[(df['target'] == setosa) | (df['target'] == versicolor)].reset_index(drop=True)
setosa_virginica_df = df[(df['target'] == setosa) | (df['target'] == virginica)].reset_index(drop=True)
versicolor_virginica_df = df[(df['target'] == versicolor) | (df['target'] == virginica)].reset_index(drop=True)
print(setosa_versicolor_df['target'].unique(), setosa_virginica_df['target'].unique(),
      versicolor_virginica_df['target'].unique())

setosa_versicolor_df.head(10)
```
